```python
# --- Day 02: Exploratory Data Analysis / Analyse exploratoire ---
# Author: Astrid Villalobos
# -----------------------------------------------

import pandas as pd
import matplotlib.pyplot as plt

# 📂 Load dataset again (from GitHub)
url = "https://raw.githubusercontent.com/astridcvr/daily-ml-practice/main/sales_data_sample.csv"
df = pd.read_csv(url, encoding='latin1')

# 🔍 Preview
df.head()
```

```python
print(df.columns.tolist())
```

```python
# Sales by Country / Ventes par pays
sales_by_country = df.groupby("COUNTRY")["PRICEEACH"].sum().sort_values(ascending=False)
print(sales_by_country)
```

```python
# Visualization / Visualisation
plt.figure(figsize=(10,5))
sales_by_country.plot(kind='bar')
plt.title("Total Sales by Country / Ventes totales par pays")
plt.ylabel("Sales ($)")
plt.xlabel("Country")
plt.tight_layout()
plt.show()
```

```python
# Sales by Product Line / # Ventes totales par ligne de produit


sales_by_productline = df.groupby("PRODUCTLINE")["SALES"].sum().sort_values(ascending=False)
sales_by_productline.plot(kind="bar", figsize=(8,5))
plt.title("Total Sales by Product Line")
plt.xlabel("Product Line")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# Order Status Distribution / Répartition du statut des commandes
status_counts = df["STATUS"].value_counts()
status_counts.plot(kind="bar", figsize=(10,5))
plt.title("Order Status Distribution / Répartition du statut des commandes")
plt.xlabel("Order Status")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Sales by Deal Size / Ventes par taille d'affaire
sales_by_dealsize = df.groupby("DEALSIZE")["SALES"].sum().sort_values(ascending=False)
sales_by_dealsize.plot(kind="bar", figsize=(6,4))
plt.title("Sales by Deal Size / Ventes par taille d'affaire")
plt.xlabel("Deal Size / Taille d'affaire")
plt.ylabel("Total Sales / Ventes totales")
plt.show()
```
